# **Summary 2**

As we explore the dataset we come to know that some columns are not assigned the correct data types and there are some null values also. So we cleaned the data and fill or delete the null values according to the number of null values in the respective columns of the dataset given.  

**Assigning** **correct datatypes to the columns of the dataset.**

```python
play_store_data.info()
```

```python
# Changing the data type of Size column from object to float
def size(size):
  if 'k' in str(size):
    size = size.replace('k','')
    size=float(size)
    size=size/1024
  elif 'M' in str(size):
    size=size.replace('M','')
    size=float(size)
  elif 'Varies with device' in str(size):  
    size=size.replace('Varies with device','0')
    size = float(size)
  return size

play_store_data['Size']=play_store_data['Size'].apply(lambda x: size(x))

play_store_data.drop(index=play_store_data[play_store_data['Size']=='1,000+'].index,axis=0,inplace=True)
play_store_data['Size']=play_store_data['Size'].astype(float)
```

```python
# Changing the data type of Install column from object to int
def install(install):
  ''' '''
  if ',' in str(install):
    install=install.replace(',','')
  if '+' in str(install):
    install=install.replace('+','')
  if 'Free' in str(install):
    install=install.replace('Free','0')
    install=float(install)
    
  return install

play_store_data['Installs']=play_store_data['Installs'].apply(lambda x: install(x))
play_store_data['Installs']=play_store_data['Installs'].astype(float)
```

```python
# Changing the data type of Price column from object to float
def price(price):
  ''' '''
  if '$' in str(price):
    price=price.replace('$','')
    price=float(price)
  if 'Everyone' in str(price):
    price=price.replace('Everyone','0')
    price=float(price)
    
  return price

play_store_data['Price']=play_store_data['Price'].apply(lambda x: price(x))
play_store_data['Price']=play_store_data['Price'].astype(float)
```

```python
# Changing the data type of Reviews column from object to float
def reviews(reviews):
  ''' '''
  if 'M' in str(reviews):
    reviews =reviews.replace('M','')
    reviews=float(reviews)*1000000
  else:
    reviews=reviews
    return reviews

play_store_data['Reviews']=play_store_data['Reviews'].apply(lambda x: reviews(x))
play_store_data['Reviews']=play_store_data['Reviews'].astype(float)
```

```python
play_store_data.info()
```

**Dealing with the null values**

```python
play_store_data.isnull().sum()
```

```python
# There are more than 10% of values(1474) in the Rating column null so filling them with the mean of the rating values
play_store_data['Rating'].fillna(np.mean(play_store_data['Rating']),inplace=True)
play_store_data['Rating'].isnull().sum()
```

```python
# There is one value null in Type column so deleting it 
play_store_data.drop(index=play_store_data[play_store_data['Type'].isna()].index,axis=0,inplace=True)
play_store_data['Type'].isnull().sum()
```

```python
# There is 8 values null in Current Ver column so deleting it 
play_store_data.drop(index=play_store_data[play_store_data['Current Ver'].isna()].index,axis=0,inplace=True)
play_store_data['Current Ver'].isnull().sum()
```

```python
# There is 2 values null in Android Ver column so deleting it 
play_store_data.drop(index=play_store_data[play_store_data['Android Ver'].isna()].index,axis=0,inplace=True)
play_store_data['Android Ver'].isnull().sum()
```

# **Insights of the dataset** 

**Q 1. Which are the categories of app having highest number of installs?**

```python
# extracting sum of the installs for each category
category_installs=play_store_data.groupby('Category')['Installs'].sum()
category_installs

```

```python
# Creating the Bar Plot
plt.figure(figsize=(10,6))

category_installs.plot(kind='bar')
plt.title('INSTALLS FOR EACH CATEGORY',size =20)
plt.xlabel('Category')
plt.ylabel('Number of Installs')
plt.show()
```

**Inference**

From the bar plot shown above we can see that most number of apps are of category:- Game, Communication, Productivity, Social and Tools.

**Q 2 What is the rating of most number of apps?**

```python
# Applying the round function(upto 1 decimal place) on the rating column values
play_store_data['Rating']=round(play_store_data['Rating'],1)
```

```python
# Extracting number of apps for each rating
rating_no_of_apps=play_store_data.groupby('Rating')['App'].count()
rating_no_of_apps
```

```python
# Creating the Bar Plot
plt.figure(figsize=(10,6))

rating_no_of_apps.plot(kind='bar')
plt.title('')
plt.xlabel('Rating')
plt.ylabel('Number of Apps')
plt.show()
```

**Inference**

From the above bar plot shown most number of apps are having rating from  4.1 to 4.6.

**Q 3 Which categories got highest average number of reviews?**

```python
# Extracting mean of the reviews for each category
category_reviews=play_store_data.groupby('Category')['Reviews'].mean()
category_reviews.reset_index()
category_reviews
```

```python
# Creating the Bar Plot
plt.figure(figsize=(10,6))

category_reviews.plot(kind='bar')
plt.title('MEAN OF THE REVIEWS FOR EACH CATEGORY')
plt.xlabel('Category')
plt.ylabel('Reviews')
plt.show()
```

**Inference**

From the above bar plot shown Communication, Game, Social, Photography and Video players Categories got most number of average of reviews.

**Q 4 What is the effect of size of an app on type and rating?**

From the above line plot we come to know that apps having high ratings are of greater sizes.

**Q 5 What is the effect of price of an app on rating?**

**Q 6 What is the share of each of the content rating in play store?**

```python
# Creating the Pie Chart
plt.figure(figsize=(25,10))
palette_color=sns.color_palette('deep')
labels=['Everyone','Teen','Everyone 10+','Mature 17+','Adults only 18+','Unrated']
data_frame=pd.DataFrame(play_store_data['Content Rating'].value_counts())
plt.pie(data=data_frame,labels=labels,x='Content Rating',colors=palette_color,autopct='%.0f%%')
plt.legend(labels)
plt.show()
```

**Inference**

From the pie chart we come to know that majority(80%) of the apps are for everyone in the play store.

**Q 7 What is the relation among the price,size,rating, reviews and installs of apps?**

```python
# creating the correlation matrix
corr_df=play_store_data[['Installs','Price','Size','Rating','Reviews']]
corr_df=corr_df.corr()
corr_df
```

```python
# Creating the Heatmap
plt.figure(figsize=(10,6))
plt.title('HEATMAP',size=20)
sns.heatmap(corr_df,vmin=-1,cmap='coolwarm',annot=True)
```

**Inference**

From the correlation matrix and heatmap we come to know that there is not any considerable correlation among the features except      
reviews and installs(that is 0.64).
